```python
dfAG = pd.read_csv('DimJiaLiAutoGrad.csv', index_col=0)
dfAG
```

```python
dfAN = pd.read_csv('DimJiaLiAnalytical.csv', index_col=0)
dfAN
```

```python
dfEXP = pd.read_csv('ExperimentalDimJiaLi.csv', index_col = 0).reset_index()
dfEXP['Pc'] = dfEXP['Pc']*1e3
```

```python
dfErr = pd.concat([(dfAG.Tc-dfEXP.Tc)/dfEXP.Tc, (dfAG.Pc-dfEXP.Pc)/dfEXP.Pc,\
                   (dfAN.Tc-dfEXP.Tc)/dfEXP.Tc, (dfAN.Pc-dfEXP.Pc)/dfEXP.Pc], axis = 1)
dfErr.columns = ['Tc AG', 'Pc AG', 'Tc AN', 'Pc AN']
dfErr['Tdev'] = dfErr['Tc AG']-dfErr['Tc AN']
dfErr['Pdev'] = dfErr['Pc AG']-dfErr['Pc AN']
dfErr
```

```python
dfnF = pd.concat([dfAG.iter, dfAG.Fev, dfAN.iter, dfAN.Fev], axis = 1)
dfnF['diter'] = dfnF.iloc[:, 0] - dfnF.iloc[:, 2] 
dfnF['dFev'] = dfnF.iloc[:, 1] - dfnF.iloc[:, 3] 
dfnF
```

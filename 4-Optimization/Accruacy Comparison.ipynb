```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

```python
MxN = 4

df = pd.read_csv("TrainingSets/MX_"+str(MxN)+"_Train.csv", index_col = 0)
df
```

```python
df['Tc_OPT'] = df['Tc_OPT'].replace(0, np.nan)
df['Pc_OPT'] = df['Pc_OPT'].replace(0, np.nan)
```

```python
fig, ax = plt.subplots(1, 2, figsize = [10, 5], sharex = True, sharey = True)
ax1 = ax[0]
ax2 = ax[1]
ax11 = ax1.twinx()
ax22 = ax2.twinx()

plt.subplots_adjust(wspace = 0)

ax1.set_title('Dampened Newton Raphson')
ax1.plot(df.iloc[:, 0], df['Pc_NR'], c = 'b')
ax11.plot(df.iloc[:, 0], df['Tc_NR'], c = 'r')
ax11.set_yticks([])
ax1.set_ylabel('Pressure (Pa)')

ax2.set_title('Differential Evolution')
ax2.plot(df.iloc[:, 0], df['Pc_OPT'], c = 'b')
ax22.plot(df.iloc[:, 0], df['Tc_OPT'], c = 'r')
ax22.set_ylabel('Temperature (K)')
ax22.set_ylim(ax11.get_ylim())

ax1.set_xlabel(df.columns[0]+' Fraction')
ax2.set_xlabel(df.columns[0]+' Fraction')

ax1.legend(['Pressure'])
ax22.legend(['Temperature'])
```
